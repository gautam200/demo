**KNeighbors Classifier**

```python
import pandas as pd                                       #data manipulation
import numpy as np                                        #data manipulation
import sklearn                                            #ML support
from sklearn.neighbors import KNeighborsClassifier as knn #KNN Classifier
```

```python
df = pd.read_csv("knn.csv")                             #Loading the data into data frame
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,2].values
```

```python
x
```

```python
y
```

```python
classifier = knn(n_neighbors=3)         #creating a knn model
```

```python
classifier.fit(x,y)                     #training the data
```

```python
x_test= np.array([7,6])
y_predict = classifier.predict([x_test])        #predicting using model
```

```python
y_predict
```
